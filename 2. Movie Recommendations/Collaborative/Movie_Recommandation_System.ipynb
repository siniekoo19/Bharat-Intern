```python
import pandas as pd
import numpy as np 
```

```python
ratings  = pd.read_csv('dataset/ratings.csv')
ratings
```

```python
movies  = pd.read_csv('dataset/movies.csv')
movies
```

```python
dataset = movies.merge(ratings, on = 'movieId')[['movieId', 'title', 'userId', 'rating']]
dataset.columns = ['MovieId', 'MovieTitle', 'UserId', 'Rating']
dataset.set_index('MovieId', inplace = True)
dataset
```

```python
# Finding the user having total rating 200 or more
x = dataset.groupby('UserId').count()['Rating'] > 200
user_rating_200 = x[x].index
```

```python
# Filtered dataset where only which users are present who gave total rating 200 or more
filtered_dataset = dataset[dataset['UserId'].isin(user_rating_200)]
filtered_dataset
```

```python
# Finding the famous movie having ratings of atleast 50
y =  dataset.groupby('MovieTitle').count()['Rating'] > 50
famous_movies = y[y].index
```

```python
final_dataset = filtered_dataset[filtered_dataset['MovieTitle'].isin(famous_movies)]
final_dataset
```

```python
# Creating the pivot table 
pt = final_dataset.pivot_table(index='MovieTitle', columns='UserId', values='Rating')
pt.fillna(value = 0, inplace=True)
pt
```

```python
import pickle
pickle.dump(pt.to_dict(), open('movies_dict.pkl','wb'))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(pt)
similarity # similarity score of each movie wrt another
```

```python
pickle.dump(similarity,open('similarity.pkl','wb'))
```

```python
movies_dict = pickle.load(open('movies_dict.pkl', 'rb'))
movies_df = pd.DataFrame(movies_dict)
similarity = pickle.load(open('similarity.pkl', 'rb'))


def recommand(movie_name):
    index = np.where(movies_df.index == movie_name)[0][0]
    similar_items = sorted(list(enumerate(similarity[index])), key = lambda x : x[1], reverse = True)[1:6]
    
    ia = IMDb()

    recommended_movies = []
    recommended_movies_posters = []
    
    for i in similar_items:
        # Similar Movie
        recommended_movies.append(movies_df.index[i[0]])
        # Similar Movie Poster
        m = ia.search_movie(movies_df.index[i[0]])
        result = m[0]
        # poster = dict(result)["full-size cover url"]
        recommended_movies_posters.append(result.keys())

    return recommended_movies, recommended_movies_posters
```
